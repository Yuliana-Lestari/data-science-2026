```python
import numpy as np

# Buat random generator
rng = np.random.default_rng(42)

# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

print("a + b:", a + b)
print("a * b:", a * b)
print("a^2:", a ** 2)

# Universal functions (ufuncs)
print("\nAkar:", np.sqrt(a))
print("Exp:", np.exp(a))
print("Log:", np.log(b))

# Fungsi statistik
data = rng.normal(0, 1, 1000)

print("\nMean:", round(np.mean(data), 4))
print("Std :", round(np.std(data), 4))
print("Min :", round(np.min(data), 4))
print("Max :", round(np.max(data), 4))
print("Q1  :", round(np.percentile(data, 25), 4))
print("Q3  :", round(np.percentile(data, 75), 4))
```

```python
# Python — Boolean Indexing
import pandas as pd

# Load dataset Titanic (URL lengkap)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Filter satu kondisi
penumpang_selamat = df[df['Survived'] == 1]
penumpang_wanita  = df[df['Sex'] == 'female']

print("Jumlah penumpang selamat:", len(penumpang_selamat))
print("Jumlah penumpang wanita:", len(penumpang_wanita))

# Filter kondisi ganda (& = AND, | = OR)
wanita_kelas1 = df[(df['Sex']=='female') & (df['Pclass']==1)]

print("\nWanita kelas 1:")
print(wanita_kelas1.head())

# Seleksi kolom tertentu
print("\nKolom tertentu:")
print(df[['Name','Age','Sex','Survived']].head())
```

```python
import pandas as pd

# Pastikan dataset sudah ada
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Pola dasar: groupby → aggregation
print(df.groupby('Pclass')['Survived'].mean())

# Agregasi multiple kolom & fungsi
hasil = df.groupby('Pclass').agg(
    total      = ('PassengerId', 'count'),
    selamat    = ('Survived', 'sum'),
    pct_selamat= ('Survived', 'mean'),
    rata_usia  = ('Age', 'mean'),
    max_fare   = ('Fare', 'max')
)

hasil = hasil.round(3)
print("\nHasil agregasi:")
print(hasil)

# GroupBy multi-kolom (diperbaiki tanpa backslash)
multi = df.groupby(['Pclass','Sex'])['Survived'].mean().unstack().round(3)

print("\nGroupBy Pclass & Sex:")
print(multi)
```
