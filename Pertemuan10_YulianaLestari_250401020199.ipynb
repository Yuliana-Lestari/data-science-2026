```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report, roc_auc_score
```

```python
df = pd.read_csv("telco_churn.csv")
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df["Churn"].value_counts())
```

```python
print(df["Churn"].value_counts(normalize=True))
```

```python
df = df.drop("customerID", axis=1)
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
df = df.dropna()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop("Churn_Yes", axis=1)

y = df["Churn_Yes"]
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(

    X,

    y,

    test_size=0.2,

    stratify=y,

    random_state=42

)
```

```python
rf = RandomForestClassifier(

    n_estimators=300,

    class_weight="balanced",

    random_state=42

)
```

```python
_ = rf.fit(X_tr, y_tr)
```

```python
pred = rf.predict(X_te)
```

```python
proba = rf.predict_proba(X_te)[:,1]
```

```python
print(classification_report(y_te,pred))
```

```python
print("ROC-AUC :",roc_auc_score(y_te,proba))
```

```python
hasil = pd.DataFrame({

    "Probabilitas Churn":proba

})

hasil.head()
```

# Kesimpulan

Model Random Forest berhasil digunakan untuk memprediksi customer churn dengan akurasi sebesar **79%** dan nilai **ROC-AUC sebesar 0,820**. Model memiliki kemampuan yang cukup baik dalam membedakan pelanggan yang berpotensi churn dan tidak churn. Meskipun demikian, nilai recall pada kelas churn masih dapat ditingkatkan agar lebih banyak pelanggan berisiko dapat terdeteksi. Hasil probabilitas dari `predict_proba()` dapat dimanfaatkan perusahaan untuk membantu menentukan strategi retensi pelanggan.
