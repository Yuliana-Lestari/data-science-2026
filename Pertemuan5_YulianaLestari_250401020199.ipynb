```python
# =========================
# IMPORT LIBRARY
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

# =========================
# LOAD DATASET
# =========================
df = sns.load_dataset('tips')

# Melihat data
print(df.head())

# Informasi dataset
print(df.shape)
print(df.info())

# =========================
# SET STYLE
# =========================
sns.set_theme(style='whitegrid')

# =========================
# MEMBUAT DASHBOARD
# =========================
fig = plt.figure(figsize=(16,12))

fig.suptitle(
    'Dashboard Analisis Dataset Tips',
    fontsize=18,
    fontweight='bold'
)

# Layout 2x2
gs = gridspec.GridSpec(2,2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

# =========================
# 1. BAR CHART
# =========================
avg_by_day = df.groupby('day')['total_bill'].mean()

avg_by_day.plot(
    kind='bar',
    ax=ax1,
    color='skyblue'
)

ax1.set_title('Rata-rata Total Bill per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Total Bill')

# =========================
# 2. HISTOGRAM
# =========================
sns.histplot(
    data=df,
    x='total_bill',
    kde=True,
    color='orange',
    ax=ax2
)

ax2.set_title('Distribusi Total Bill')

# =========================
# 3. BOXPLOT
# =========================
sns.boxplot(
    data=df,
    x='day',
    y='total_bill',
    palette='Set2',
    ax=ax3
)

ax3.set_title('Distribusi Total Bill per Hari')

# =========================
# 4. SCATTER PLOT
# =========================
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    palette='Set1',
    ax=ax4
)

ax4.set_title('Hubungan Total Bill dan Tip')

plt.tight_layout()

# =========================
# SAVE GAMBAR
# =========================
plt.savefig('dashboard_tips.png')

plt.show()

print("Dashboard berhasil dibuat!")
```

### Analisis Bar Chart

What?
Grafik menunjukkan rata-rata total bill pelanggan pada setiap hari. Hari Minggu memiliki rata-rata total bill tertinggi dibanding hari lainnya.

So what?
Hal ini menunjukkan bahwa pelanggan cenderung mengeluarkan lebih banyak uang saat akhir pekan. Kemungkinan karena pelanggan datang bersama keluarga atau teman.

Now what?
Restoran dapat menyiapkan lebih banyak staf dan persediaan makanan pada akhir pekan untuk mengantisipasi peningkatan pengeluaran pelanggan.

### Analisis Histogram

What?
Distribusi total bill menunjukkan sebagian besar pelanggan memiliki total bill di kisaran 10 hingga 20 dolar.

So what?
Hal ini menandakan bahwa mayoritas transaksi pelanggan berada pada kategori menengah dan hanya sedikit pelanggan dengan total bill sangat tinggi.

Now what?
Restoran dapat membuat promosi khusus untuk meningkatkan pembelian pelanggan agar total bill meningkat.

### Analisis Boxplot

What?
Grafik boxplot menunjukkan persebaran total bill pada tiap hari dan terdapat beberapa outlier terutama pada hari Minggu.

So what?
Hal ini berarti terdapat beberapa pelanggan yang melakukan pembayaran jauh lebih tinggi dibanding pelanggan lainnya.

Now what?
Perlu dianalisis lebih lanjut apakah pembayaran tinggi tersebut berasal dari acara khusus atau jumlah pelanggan yang lebih besar.

### Analisis Scatter Plot

What?
Scatter plot menunjukkan hubungan positif antara total bill dan tip. Semakin besar total bill, semakin besar tip yang diberikan.

So what?
Hal ini menunjukkan pelanggan cenderung memberikan tip berdasarkan jumlah pembayaran mereka.

Now what?
Restoran dapat meningkatkan kualitas pelayanan agar pelanggan lebih terdorong memberikan tip lebih besar.
